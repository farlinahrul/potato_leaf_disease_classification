```python
from rembg.bg import remove
import numpy as np
import io
from PIL import Image
```

```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
import os
```

```python
path = r"D:\PROJECTS\Data Science\PROJECTS\potato\dataset"
new_path = "D:\\PROJECTS\\Data Science\\PROJECTS\\potato\\new dataset\\"

fill_color = (255,255,255) 

for dirname in os.listdir(path):
    x = 1
    dirpath = os.path.join(path, dirname)
    print(dirpath)
    for filename in os.listdir(dirpath):
        f = np.fromfile(os.path.join(dirpath, filename))
        print(filename)
        result = remove(f)
        img = Image.open(io.BytesIO(result)).convert("RGBA")
        img = img.resize((128,128)) # Resize Image Dataset
        
        if img.mode in ('RGBA', 'LA'):
            background = Image.new(img.mode[:-1], img.size, fill_color)
            background.paste(img, img.split()[-1]) 
            img = background
            
        # Renaming File
        if (x < 10):
            img.convert("RGB").save(new_path+dirname+"\\"+'00'+str(x)+'.jpg')
        elif (x < 100):
            img.convert("RGB").save(new_path+dirname+"\\"+'0'+str(x)+'.jpg')
        else:    
            img.convert("RGB").save(new_path+dirname+"\\"+str(x)+'.jpg')
        x = x + 1
```
